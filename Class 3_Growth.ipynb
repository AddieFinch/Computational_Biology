# Growth

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# EXPONENTIAL GROWTH MODEL

nsteps = 100 #number of steps in our simulation
x = np.zeros(nsteps) #number of animals at every step
x[0] = 1 #the original number of animals
r = 1.1 #rate of pop change: (growth - die out??)... if r is above 1, pop grows. below 1, pop die out

for i in range(1, nsteps):
    x[i] = x[i-1]*r
    plt.plot(x);
```

```python
#LOGISTIC GROWTH MODEL

nsteps = 100 #number of steps in our simulation
x = np.zeros(nsteps) #number of animals at every step
x[0] = 0.6 #the original "population size"
r = 3.4 #rate of pop change: (growth - die out??)... if r is above 1, pop grows. below 1, pop die out

for i in range(1, nsteps):
    x[i] = r*x[i-1]*(1-x[i-1])

plt.plot(x);
plt.ylim((0,1)) #Force y axis to be between 0 and 1
```
