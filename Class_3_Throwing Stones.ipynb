# Simple Numerical Integration for Motion

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#ATTEMPT 1: THROWING STONE UPWARDS
a = -1       #acceleration; negative because it tries to reverse the flight of the stone
v = 5        #we throw the stone upward, so it is positive
x = 0        #our position; we throw it from the surface
dt = 0.01    #our step --> numerical integration step supposed to be small
nsteps = 1000
history = np.zeros(nsteps) #We will plot this eventually #Creating 1000 zeros and replacing them with normal numbers
for i in range(nsteps):
    v = v + a*dt
    x = x + v*dt
    history[i] = x
    
plt.plot(history);
```

```python
#ATTEMPT 2: THROWING IT ON JUPITER
a = -2       #what if we're throwing stones on jupiter -- strong gravity -- this is a parameter of the model
v = 5        #we throw the stone upward, so it is positive
x = 0        #our position; we throw it from the surface
dt = 0.01    #our step --> numerical integration step supposed to be small
nsteps = 1000
history = np.zeros(nsteps) #We will plot this eventually #Creating 1000 zeros and replacing them with normal numbers
for i in range(nsteps):
    v = v + a*dt
    x = max(x + v*dt, 0)
    history[i] = x
    
plt.plot(history);
```

```python
#ATTEMPT 3: THROWING AT AN ANGLE
a = -1       
vy = 5 #Vertical component of the speed, and it is upwards --> positive
vx = 1 #Horizontal component of the speed  
x = 0  
y = 0
dt = 0.01 #can make 
nsteps = round(10/dt)

for i in range(nsteps):
    vy = vy + a*dt #update vertical speed
    
    x = x + vx*dt #update x position
    y = max(y + vy*dt , 0) #update y position
    history[i] = x
    
    plt.plot(x,y, 'k.'); #Just plot a new point instead of remembering x and y bc we're lazy rn
    
# will need to calculate it with air friction
# we don't have it for horizontal speed -- need to set it as a constant and add it to one row
```

```python
#ATTEMPT 4: THROWING AT AN ANGLE AND ADDING FRICTION -- THROWING A HAMBURGER
a = -1       
vy = 5 #Vertical component of the speed, and it is upwards --> positive
vx = 1 #Horizontal component of the speed 
x = 0   
y = 0
f = 0.01 #this is friction -- it is supposed to stay constant -- affects speed #how important friction is relative to gravity
dt = 0.01 

nsteps = round(10/dt)

for i in range(nsteps):
    vy = vy + a*dt - vy*f
    vx = vx - vx*f
    x = x + vx*dt #update x position
    y = max(y + vy*dt , 0) #update y position
    history[i] = x
    plt.plot(x,y,'k.'); #Just plot a new point instead of remembering x and y bc we're lazy rn
```
