```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
%matplotlib notebook

n_animals = 50
n_steps = 50
diffusion_speed = 0.03
speed = 0.05

x = np.ones(n_animals) * 0.5
y = np.ones(n_animals) * 0.5

fig = plt.figure(figsize=(3,3))
plt.xlim((0,1));
plt.ylim((0,1));
 
for t in range(n_steps):
    x += np.random.randn(n_animals)*diffusion_speed
    y += np.random.randn(n_animals)*diffusion_speed

    for a in range(n_animals):
        
        dx = np.random.rand(n_animals)
        dy = np.random.rand(n_animals)
        x += dx * speed
        y += dy * speed
        
        x = np.where(x>0, x, dx) #This is how borders were implemented
        x = np.where(x<1, x, dx)
        y = np.where(y>0, y, dy)
        y = np.where(y<1, y, dy)

        plt.cla()
        plt.plot(x,y,'k.',markersize=2);
        #time.sleep(0.01)
        plt.xlim((0,1));
        plt.ylim((0,1));
        fig.canvas.draw()
    
x = np.maximum(x,1)
x = np.minimum(x,0)
y = np.maximum(y,1)
y = np.minimum(y,0)
```
