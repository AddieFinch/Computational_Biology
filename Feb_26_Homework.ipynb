```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
%matplotlib notebook

n_animals = 50
n_steps = 50
diffusion_speed = 0.03
speed = 0.05

x = np.ones(n_animals) * 0.5
y = np.ones(n_animals) * 0.5

vx = np.random.rand(n_animals)
vy = np.random.rand(n_animals)

fig = plt.figure(figsize=(3,3))
plt.xlim((0,1));
plt.ylim((0,1));
 
for t in range(n_steps):
    
    x += np.random.randn(n_animals)*diffusion_speed
    y += np.random.randn(n_animals)*diffusion_speed
    
    for a in range(n_animals):
        
        x = x + vx * speed
        y = y + vy * speed
        
       #Tried implementing borders with an if statement, but I kept getting errors, so I went the np.where route instead.
       # if x<0:
       #     x = -vx
       #  if y<0:
       #     y = -vy
       # if x>1:
       #     x = -vx
       # if y>1:
       #     y = -vy
        
        x = np.where(x>0, x, vx) #Borders?
        x = np.where(x<1, x, vx)
        y = np.where(y>0, y, vy)
        y = np.where(y<1, y, vy)

        plt.cla()
        plt.plot(x,y,'k.',markersize=2);
        #time.sleep(0.01)
        plt.xlim((0,1));
        plt.ylim((0,1));
        fig.canvas.draw()
    
x = np.maximum(x,1)
x = np.minimum(x,0)
y = np.maximum(y,1)
y = np.minimum(y,0)
```
