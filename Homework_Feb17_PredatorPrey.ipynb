```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#EQUAL STARTING VALUES OF BUNNIES AND FOXES

nsteps = 20000
a = 1.05 #rate at which the bunnies reproduce
b = 1 #rate at which bunnies die 
c = 1.5 #how many bunnies are equal to one fox
d = 1 #rate at which foxes die


dt = 0.001 #steppy step

possible_ds = [0.5, 1, 1.5, 2.0]  

x = np.zeros(nsteps)
y = np.zeros(nsteps)

fig1 = plt.figure(figsize=(16,5));
fig2 = plt.figure(figsize=(5,5));
ax2 = plt.axes()

for itrial in range(len(possible_ds)):
    d = possible_ds[itrial] 
    x[0] = 1 
    y[0] = 1

    for t in range(1,nsteps):
        x[t] = x[t-1] + dt*(a*x[t-1] - b*x[t-1]*y[t-1])
        y[t] = y[t-1] + dt*(c*x[t-1]*y[t-1] - d*y[t-1])

    ax1 = fig1.add_subplot(1,len(possible_ds),itrial+1)
    ax1.plot(x, label='Bunnies');
    ax1.plot(y, label='Foxes');
    ax1.legend(loc='upper right');

    ax2.plot(x,y, label=d);
    plt.xlabel('Bunnies');
    plt.ylabel('Foxes');
 
ax2.legend(loc='upper right');
```

```python
#TRIPLE THE STARTING VALUE OF BUNNIES PER FOXES

nsteps = 20000
a = 1.05 #rate at which the bunnies reproduce
b = 1 #rate at which bunnies die 
c = 1.5 #how many bunnies are equal to one fox
d = 1 #rate at which foxes die


dt = 0.001 #steppy step

possible_ds = [0.5, 1, 1.5, 2.0]  

x = np.zeros(nsteps)
y = np.zeros(nsteps)

fig1 = plt.figure(figsize=(16,5));
fig2 = plt.figure(figsize=(5,5));
ax2 = plt.axes()

for itrial in range(len(possible_ds)):
    d = possible_ds[itrial]
    x[0] = 3
    y[0] = 1

    for t in range(1,nsteps):
        x[t] = x[t-1] + dt*(a*x[t-1] - b*x[t-1]*y[t-1])
        y[t] = y[t-1] + dt*(c*x[t-1]*y[t-1] - d*y[t-1])

    ax1 = fig1.add_subplot(1,len(possible_ds),itrial+1)
    ax1.plot(x, label='Bunnies');
    ax1.plot(y, label='Foxes');
    ax1.legend(loc='upper right');

    ax2.plot(x,y, label=d);
    plt.xlabel('Bunnies');
    plt.ylabel('Foxes');
 
ax2.legend(loc='upper right');
```

Here we can see how varying starting position/values of foxes and bunnies (i.e., x[0] and y[0] impacts population oscillations at various values of d (aka, fox death rate). (At least I'm pretty sure that this is what this represents. Plz correct me if I'm mistaken!)

In the first cell of code, I set x[0] and y[0] equal to each other. We can see that the fox population completely overruns the bunny population when d = 0.5 and 1 with pretty severe fluctuations, but then when the death rate is equal to 1.5, the fluctuations are much smaller, but the fox population is still greater than the bunny population. However, when d = 2.0, the bunnies are finally able to have their victory over the foxes, who are likely dying too quickly to make a large dent in the bunny population. Looking at the phase plot, you can see these fluctuations pretty well. Particularly cool is the 1.5 value, as it isn't as obvious in the "regular" plots how much the fluctuations decrease at a d of 1.5. 

In the second cell of code, I set x[0] = 1 and y[0] = 3 in order to simluate what would happen with a much greater starting value of foxes. A greater number of foxes actually allows the bunny population to catch up a bit with the fox population, but not completely overrun it. I think this is most likely due to the fact that there are so many foxes and not enough bunnies to sustain the foxes when the death rate of foxes is increased. The oscillations are way more even across values of d when the number of foxes is increased.

In the third cell, I set x[0] = 3 and y[0] = 1.
